```python
cities1=['va/aldie','va/aldie','va/leesburg', 'md/fredrick', 'md/fredrick','md/hagerstown', 'md/hagerstown','pa/breezewood', 'pa/breezewood', 'pa/altoona', 'pa/ridgway', 'pa/bradford', 'ny/ellicottville','ny/ellicottville', 'ny/springville','ny/buffalo', 'ny/buffalo']
time1=[  '7:00 am', '8:00 am', '8:00 am',     '8:00 am',     '9:00 am',     '8:00 am',       '9:00 am',      '9:00 am',      '10:00 am',       '10:00 am',   '12:00 am',   '1:00 pm',     '1:00 pm',          '2:00 pm',         '2:00 pm',       '2:00 pm',    '3:00 pm',]
```

```python
cities2=['ny/blasdell', 'ny/dunkirk','pa/erie', 'oh/ashtabula','oh/cleaveland','oh/huron','oh/toledo','mi/monroe','mi/monroe','mi/canton','mi/wixom','mi/wixom']
time2=[  '3:00 pm',      '4:00 pm',   '5:00 pm',  '6:00 pm',    '7:00 pm',      '8:00 pm', '9:00 pm',  '9:00 pm',  '10:00 pm', '10:00 pm', '10:00 pm', '11:00 pm']
```

```python
cities=cities1+cities2
time=time1+time2
len(cities)
```

```python
url1 = 'https://www.wunderground.com/hourly/us/'
url2 = '/date/2021-1-2'
driver = webdriver.Chrome()
data=[]

for i,j in zip(cities,time):
    
    url = url1+i+url2    
    driver.get(url)
    tables = WebDriverWait(driver,20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table")))
    for table in tables:
        newTable = pd.read_html(table.get_attribute('outerHTML'))
        
    x=newTable[0].fillna('')
    x=x.drop(x.index[len(x)-1])
    x=x.set_index('Time')
    x=x.loc[j]
    data.append(x)
```

```python
data=pd.DataFrame(data)
```

```python
location=[]
for i in cities:
    location.append(i.split('/')[-1])
```
